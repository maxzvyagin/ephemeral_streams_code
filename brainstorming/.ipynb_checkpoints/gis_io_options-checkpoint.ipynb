```python
import rasterio
```

```python
# defining a generator function which will yield chunks of the image at a time
def get_image_chunk(filename):
    with rasterio.open(filename) as src:
        for ji, window in src.block_windows():
            chunk = src.read()
            yield chunk
```

```python
get_image_chunk("/Users/mzvyagin/Documents/GISProject/nucleus_data/Ephemeral_Channels/Imagery/vhr_2012_refl.img")
```
