```python
### this block allows for writing the mask to a .tif file which can be opened in qGIS in order to visually check accuracy
with rasterio.Env():

    # Write an array as a raster band to a new 8-bit file. For
    # the new file's profile, we start with the profile of the source
    profile = twelve.profile

    # And then change the band count to 1, set the
    # dtype to uint8, and specify LZW compression.
    profile.update(
        dtype=rasterio.uint8,
        count=1,
        compress='lzw')

    with rasterio.open('example.tif', 'w', **profile) as dst:
        dst.write(updated_shapes_mask.astype(rasterio.uint8), 1)
```

```python
# now we want to pull out these bounding boxes
boundaries = []
for x in geom:
    b = x.envelope
    boundaries.append(b)
    
# now the we have the boundary boxes, we can convert that to a mask as well
boundary_mask = rasterio.features.geometry_mask(boundaries, w.shape, twelve.transform, all_touched=False, invert=True)
```
